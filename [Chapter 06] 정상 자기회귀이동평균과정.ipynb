# 6장 정상 자기회귀이동평균과정

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
#그림 6-1 AR(1)과정의 시계열그림 p.216
#phi=0.5 ; 양수인 경우
#Z(t-1)을 z1이라 설정하고 초기값 설정 그래프를 그리기 위해 x축과 y축의 리스트를 생성(for문으로 확장)
#p.216 그림 6-1에 있는 식을 참고
z1=0
x_list=[]
y_list=[]
for i in range(0,100):
    e=np.random.normal(0,1,1)
    z=0.5*z1+e
    x_list.append(i)
    y_list.append(z)
    z1=z
plt.figure(figsize=(12,8))
plt.plot(x_list, y_list)
plt.axhline(y=0, color='r',linestyle='--',linewidth=1)
plt.xlabel('time')
plt.ylabel('Z(t)')
plt.title('AR(1)과정의 시계열그림(phi=0.5)')
plt.show
```

```python
#phi가 양수인 경우 acf와 pacf
z1=0
x_list=[]
y_list=[]
for i in range(0,1000):
    e=np.random.normal(0,1,1)
    z=0.9*z1+e
    x_list.append(i)
    y_list.append(z)
    z1=z
plot_acf(y_list,lags=10)
plot_pacf(y_list, lags=10)
```

```python
#phi=-0.5 음수인 경우
z1=0
x_list=[]
y_list=[]
for i in range(0,100):
    e=np.random.normal(0,1,1)
    z=-0.5*z1+e
    x_list.append(i)
    y_list.append(z)
    z1=z
plt.figure(figsize=(12,8))
plt.plot(x_list, y_list)
plt.axhline(y=0, color='r',linestyle='--',linewidth=1)
plt.xlabel('time')
plt.ylabel('Z(t)')
plt.title('AR(1)과정의 시계열그림(phi=-0.5)')
plt.show
```

```python
#phi가 음수인 경우 acf와 pacf
z1=0
x_list=[]
y_list=[]
for i in range(0,1000):
    e=np.random.normal(0,1,1)
    z=-0.9*z1+e
    x_list.append(i)
    y_list.append(z)
    z1=z
plot_acf(y_list,lags=10)
plot_pacf(y_list, lags=10)
```

```python
# 그림 6-1 과 6-2를 비교했을 때 AR(1)과정에서 phi가 양수이면 음수인 경우보다 변화의 정도가 심하지 않고 부드러움을 볼 수 있다.
```

```python
#그림 6-3 (Zt-1,Zt)의 산점도 P.217
z1=0
x_list=[]
y_list=[]
for i in range(0,100):
    e=np.random.normal(0,1,1)
    z=0.5*z1+e       #phi가 양인 경우
    x_list.append(z)
    y_list.append(z1)
    z1=z
plt.figure(figsize=(12,8))
plt.plot(x_list, y_list,color='blue',linestyle='none',marker='o',markersize=5)
plt.axhline(y=0, color='r',linestyle='-',linewidth=1)
plt.axvline(x=0, color='r',linestyle='-',linewidth=1)
plt.xlabel('Z(t-1)')
plt.ylabel('Z(t)')
plt.title('(Z_t-1,Z_t)의 산점도')
plt.show
```

```python
#그림 6-3의 산점도를 보았을 때, phi가 양인 경우 Z(t-1)과 Zt는 양의 상관관계를 보이고 있음을 볼 수 있다.
```

```python
z1=0
x_list=[]
y_list=[]
for i in range(0,100):
    e=np.random.normal(0,1,1)
    z=-0.9*z1+e       #phi가 양인 경우
    x_list.append(z)
    y_list.append(z1)
    z1=z
plt.figure(figsize=(12,8))
plt.plot(x_list, y_list,color='blue',linestyle='none',marker='o',markersize=5)
plt.axhline(y=0, color='r',linestyle='-',linewidth=1)
plt.axvline(x=0, color='r',linestyle='-',linewidth=1)
plt.xlabel('Z(t-1)')
plt.ylabel('Z(t)')
plt.title('(Z_t-1,Z_t)의 산점도')
plt.show
```

```python
#phi가 음인 경우의 산점도를 보면 음의 상관이 있는 산점도가 나온다..
```
